```python
import pandas as pd
```

```python
areas = ['City of London', 'Barking and Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich', 'Hackney', 'Hammersmith and Fulham', 'Haringey', 'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington', 'Kensington and Chelsea', 'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 'Newham', 'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets', 'Waltham Forest', 'Wandsworth', 'Westminster']
```

## Population 

```python
population = pd.read_excel(open('data/population.xlsx', 'rb'),
              sheet_name='Custom age range tool', header=13)
population = population.iloc[:,1:7]
population = population.reset_index(drop = True)
```

```python
population = population[population['Area name'].isin(areas)]

population = population.drop(population.columns[[3,4]],axis =1)
population = population.groupby(['Area Code', 'Year'])['Persons'].sum().reset_index().pivot(index='Area Code', columns='Year', values='Persons')

population.index.name = 'Code'
population.to_csv('clean_data/population_annual.csv')
population
```

## Energy Consumption

```python
energy_consumption = pd.read_excel(open('data/energy consumption.xlsx', 'rb'),
              sheet_name='Data')

energy_consumption = energy_consumption[energy_consumption['Area'].isin(areas)]

energy_consumption = energy_consumption[energy_consumption['Sector'] == 'Domestic']
energy_consumption = energy_consumption[(energy_consumption['Fuel'] == 'Gas') | (energy_consumption['Fuel'] == 'Electricity')]

#for data in data
energy_consumption.reset_index(drop=True,inplace=True)

for i in range(0,energy_consumption.shape[0]):
    if energy_consumption.loc[i,'Measurement'] == 'Ktoe ':
        energy_consumption.loc[i,'Value'] = energy_consumption.loc[i,'Value'] * 10000

df = energy_consumption.groupby(['Code','Year'])['Value'].agg('sum')
df = pd.DataFrame(df)
df = df.groupby(['Code', 'Year'])['Value'].sum().reset_index().pivot(index='Code', columns='Year', values='Value')
df
result = pd.merge(energy_consumption[['Code','Area']],df,on='Code')
result = result.drop_duplicates()
result.reset_index(inplace=True,drop = True)
result.to_csv('clean_data/energy_consumption_annual.csv')
result
```
