```python
import pandas as pd
```

```python
areas = ['City of London', 'Barking and Dagenham', 'Barnet', 'Bexley', 'Brent', 'Bromley', 'Camden', 'Croydon', 'Ealing', 'Enfield', 'Greenwich', 'Hackney', 'Hammersmith and Fulham', 'Haringey', 'Harrow', 'Havering', 'Hillingdon', 'Hounslow', 'Islington', 'Kensington and Chelsea', 'Kingston upon Thames', 'Lambeth', 'Lewisham', 'Merton', 'Newham', 'Redbridge', 'Richmond upon Thames', 'Southwark', 'Sutton', 'Tower Hamlets', 'Waltham Forest', 'Wandsworth', 'Westminster']
```

## Population

```python
population = pd.read_excel(open('data/population.xlsx', 'rb'),
              sheet_name='Custom age range tool', header=13)
population = population.iloc[:,1:7]
population = population.reset_index(drop = True)

population = population[population['Area name'].isin(areas)]

population = population.drop(population.columns[[3,4]],axis =1)
#population = population.groupby(['Area Code', 'Year'])['Persons'].sum().reset_index().pivot(index='Area Code', columns='Year', values='Persons')

#population.index.name = 'Code'

#test = population[['Code','Area','2011','2012','2013','2014','2015','2016','2017']]
#df = pd.melt(test, id_vars=['Code','Area'], var_name='Year', value_name='population_annual')
df = population[(population['Year'] >= 2011) & (population['Year'] <= 2017)]
df = df.rename(columns={'Area Code':'Code','Area name':'Area'})

df.to_csv('clean_data/population_annual.csv',index=False)
df
```

## Energy Consumption

### Gas Consumption

```python
energy_consumption = pd.read_excel(open('data/energy consumption.xlsx', 'rb'),
              sheet_name='Data')

energy_consumption = energy_consumption[energy_consumption['Area'].isin(areas)]

energy_consumption = energy_consumption[energy_consumption['Sector'] == 'Domestic']
energy_consumption_electricity = energy_consumption[(energy_consumption['Fuel'] == 'Gas')]

#for data in data
energy_consumption_electricity.reset_index(drop=True,inplace=True)

for i in range(0,energy_consumption_electricity.shape[0]):
    if energy_consumption_electricity.loc[i,'Measurement'] == 'Ktoe ':
        energy_consumption_electricity.loc[i,'Value'] = energy_consumption_electricity.loc[i,'Value'] * 10000

#df = energy_consumption_electricity.groupby(['Code','Year'])['Value'].agg('sum')
#df = pd.DataFrame(df)
#df = df.groupby(['Code', 'Year'])['Value'].sum().reset_index().pivot(index='Code', columns='Year', values='Value')


energy_consumption_gas = energy_consumption_electricity[['Code','Area','Year','Value']]

energy_consumption_gas = energy_consumption_gas.rename(columns={'Value':'Gas_consumption'})

df = energy_consumption_gas[(energy_consumption_gas['Year'] >= 2011) & (energy_consumption_gas['Year'] <= 2017)]

df.to_csv('clean_data/energy_gas_consumption_annual.csv',index=False)
df
```

### Electricity Consumption

```python
energy_consumption = pd.read_excel(open('data/energy consumption.xlsx', 'rb'),
              sheet_name='Data')

energy_consumption = energy_consumption[energy_consumption['Area'].isin(areas)]

energy_consumption = energy_consumption[energy_consumption['Sector'] == 'Domestic']
energy_consumption_electricity = energy_consumption[(energy_consumption['Fuel'] == 'Electricity')]

#for data in data
energy_consumption_electricity.reset_index(drop=True,inplace=True)

for i in range(0,energy_consumption_electricity.shape[0]):
    if energy_consumption_electricity.loc[i,'Measurement'] == 'Ktoe ':
        energy_consumption_electricity.loc[i,'Value'] = energy_consumption_electricity.loc[i,'Value'] * 10000

#df = energy_consumption_electricity.groupby(['Code','Year'])['Value'].agg('sum')
#df = pd.DataFrame(df)
#df = df.groupby(['Code', 'Year'])['Value'].sum().reset_index().pivot(index='Code', columns='Year', values='Value')


energy_consumption_electricity = energy_consumption_electricity[['Code','Area','Year','Value']]

energy_consumption_electricity = energy_consumption_electricity.rename(columns={'Value':'Electricity_consumption'})

df = energy_consumption_electricity[(energy_consumption_electricity['Year'] >= 2011) & (energy_consumption_electricity['Year'] <= 2017)]

df.to_csv('clean_data/energy_electricity_consumption_annual.csv',index=False)
df
```
